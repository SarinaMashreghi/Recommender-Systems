```python
import random

NUM_CLUSTER = 10
def random_suggestion(df, prob_list, n_clusters=NUM_CLUSTER):
  clusters = np.arange(n_clusters)
  chosen_cluster = np.random.choice(clusters, p=prob_list/(prob_list.sum()))
  possible_choices = df.loc[df["cluster"]==chosen_cluster]
  recom_book = random.randint(0, len(possible_choices))
  return possible_choices.iloc[recom_book], chosen_cluster

prob = np.ones(NUM_CLUSTER)
random_suggestion(df, prob)[0]["title"], random_suggestion(df, prob)[0]["description"], random_suggestion(df, prob)[0]["cluster"]
```

##Recommendation based on description (tf-idf)

```python
from sklearn.metrics.pairwise import linear_kernel


def get_rec_tfidf(df, book, cluster):

  same_cluster_df = df.loc[df["cluster"]==cluster].reset_index()
  
  tfidf = TfidfVectorizer(stop_words='english')
  tfidf_matrix = tfidf.fit_transform(same_cluster_df['description'])

  cosine_sim_tfidf = linear_kernel(tfidf_matrix, tfidf_matrix)

  # print(same_cluster_df)
  indices = pd.Series(same_cluster_df.index, same_cluster_df['title']).drop_duplicates()
  # print(indices)

  idx = indices[book]
  # print(idx)

  sim_scores = list(enumerate(cosine_sim_tfidf[idx]))
        
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  sim_scores = sim_scores[1:11]

  movie_indices = [i[0] for i in sim_scores]

  return same_cluster_df['title'].iloc[movie_indices]

rand_book, cluster = random_suggestion(df, 20)
print(f"liked book: {rand_book['title']}") 
get_rec_tfidf(df, rand_book["title"], cluster)
```

##Recommendation based on author and category 

```python
df = df.dropna(subset=["authors"])
df
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''

features = ['authors', 'categories']

for feature in features:
    df[feature] = df[feature].apply(clean_data)
```

```python
def make_metadata(x):
    return ''.join(x['authors']) + ' ' + ''.join(x['categories']) 
df['metadata'] = df.apply((make_metadata), axis=1)
```

```python
df[["authors", "categories"]]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def get_rec_CV(df, book, cluster):

  same_cluster_df = df.loc[df["cluster"]==cluster].reset_index()
  
  count = CountVectorizer(stop_words='english')
  count_matrix = count.fit_transform(same_cluster_df['metadata'])
  # print(count_matrix)

  cosine_sim_CV = cosine_similarity(count_matrix, count_matrix)
  # print(cosine_sim_CV)

  # print(same_cluster_df)
  indices = pd.Series(same_cluster_df.index, same_cluster_df['title']).drop_duplicates()
  # print(indices)

  idx = indices[book]
  # print(idx)

  sim_scores = list(enumerate(cosine_sim_CV[idx]))
        
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  sim_scores = sim_scores[1:11]

  movie_indices = [i[0] for i in sim_scores]
 

  return same_cluster_df.iloc[movie_indices]
```
